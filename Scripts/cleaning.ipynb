```python
import pandas as pd
```

```python


cdr = pd.read_csv('../Data/cdr_data.csv')
profiles = pd.read_csv('../Data/customer_profiles.csv')
rules = pd.read_csv('../Data/rules_master.csv')
```

```python
# Clean cdr: Fill missing charges, remove bad rows, standardize dates
cdr['Charges'] = cdr['Charges'].fillna(cdr['Charges'].mean())
cdr.dropna(subset=['Customer_ID'], inplace=True)
cdr.drop_duplicates(inplace=True)
cdr['Call_Date'] = pd.to_datetime(cdr['Call_Date'], format='%d-%m-%Y')
```

```python
profiles.dropna(subset=['Customer_ID'], inplace=True)
profiles.drop_duplicates(inplace=True)
```

```python
rules['Weightage'] = rules['Weightage'].fillna(0.5)
rules.drop_duplicates(inplace=True)
```

```python
cdr.to_csv('../Data/cleaned_cdr.csv', index=False)
profiles.to_csv('../Data/cleaned_customer_profiles.csv', index=False)
rules.to_csv('../Data/cleaned_rules_master.csv', index=False)
```
